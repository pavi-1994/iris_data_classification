### Checking Prescence of outlier and removing it

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
sns.boxplot(iris_df['PetalLengthCm'])
plt.title("Box Plot for PetalLengthCm")
plt.subplot(2,2,2)
sns.boxplot(iris_df['PetalWidthCm'])
plt.title("Box Plot for PetalWidthCm")
plt.subplot(2,2,3)
sns.boxplot(iris_df['SepalWidthCm'])
plt.title("Box Plot for SepalWidthCm")
plt.subplot(2,2,4)
sns.boxplot(iris_df['SepalLengthCm'])
plt.title("Box Plot for SepalLengthCm")
```

we can observe that there is presence  outlier in the SepalWidthCm.

#### As we impute the mean value we can also impute the outlier with mean, median, zero value in this methods. Since we imputing there is no loss of data. Here median is appropriate because it is not affected by outliers.

```python
q1 = iris_df['SepalWidthCm'].quantile(0.25)
q3 = iris_df['SepalWidthCm'].quantile(0.75)
iqr = q3-q1
Lower_tail = q1 - 1.5 * iqr
Upper_tail = q3 + 1.5 * iqr
med = np.median(iris_df['SepalWidthCm'])
for i in iris_df['SepalWidthCm']:
    if i > Upper_tail or i < Lower_tail:
           iris_df['SepalWidthCm'] = iris_df['SepalWidthCm'].replace(i, med)
sns.boxplot(iris_df['SepalWidthCm'])
plt.title("Box Plot after median imputation")
plt.show()
```

### Spliting the dataset in independent and dependent variables

```python
#Spliting the dataset in independent and dependent variables
X = iris_df.drop(columns="Species")
y = (iris_df["Species"])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 82)
```

### Feature Scaling to bring the variable in a single scale


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

 #### Naive Bayes

```python
# Fitting Naive Bayes Classification to the Training set with linear kernel
from sklearn.naive_bayes import GaussianNB
nvclassifier = GaussianNB()
nvclassifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = nvclassifier.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
Accuracy_percent_test = (metrics.accuracy_score(y_test,y_pred)) * 100
print("Gaussian Naive Bayes model accuracy for test dataset: "+str(Accuracy_percent_test))
```

### Logistic regression

```python
# model
from sklearn.linear_model import LogisticRegression
Lmodel = LogisticRegression()
Lmodel.fit(X_train, y_train)
y_pred1 = Lmodel.predict(X_test)
```

```python
Accuracy_percent_test = (metrics.accuracy_score(y_test,y_pred1)) * 100
print("Logistic model accuracy for test dataset: "+str(Accuracy_percent_test))
```

### SVM

```python
# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
svm_model = SVC(gamma=0.001 , C = 100)
svm_model.fit(X_train, y_train)
y_pred2 = svm_model.predict(X_test)
```

```python
Accuracy_percent_test = (metrics.accuracy_score(y_test,y_pred2)) * 100
print("svm model accuracy for test dataset: "+str(Accuracy_percent_test))
```
